```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from torchvision import transforms
import torch
```

```python
img_path = './91.jpg'
transform = transforms.Compose([transforms.Resize((64, 96)), transforms.ToTensor()])
```

```python
img = transform(Image.open(img_path).convert('RGB'))
```

```python
img.shape
```

```python
example = torch.randn(3, 2, 2)
print(example)
example = example.permute(1, 2, 0)
print(example)
```

```python
plt.figure(1)
fig, ax = plt.subplots()
ax.imshow(img.permute(1, 2, 0).numpy())
ax.set_title("water")
ax.set_xticks([])
ax.set_yticks([])
plt.show()
```
